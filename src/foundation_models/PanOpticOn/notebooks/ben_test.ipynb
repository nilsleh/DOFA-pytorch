# Metadata Checks

```python
import pandas as pd
import os
benv2_root = '/data/panopticon/datasets/benv2'

fname = os.path.join(benv2_root, 'metadata.parquet')
metadata = pd.read_parquet(fname)
metadata
```

```python
metadata['split'].value_counts()
```

```python
metadata['patch_id'].iloc[1], metadata['s2v1_name'].iloc[1], metadata['s1_name'].iloc[1]
```

```python
s1_path = 'BigEarthNet-S1'
s2_path = 'BigEarthNet-S2'

tile_id = '_'.join(metadata['patch_id'].iloc[0].split('_')[:-2])
s1_tile_id = '_'.join(metadata['s1_name'].iloc[0].split('_')[:-3])
patch_id = metadata['patch_id'].iloc[0]
s1_name = metadata['s1_name'].iloc[0]

file_name = metadata['s1_name'].iloc[0] + '_VV.tif'
s1 = os.path.join(benv2_root, s1_path, s1_tile_id, s1_name, file_name)

print(s1)
print(os.path.exists(s1))

file_name = metadata['patch_id'].iloc[0] + '_B01.tif'
print(patch_id)

s2 = os.path.join(benv2_root, s2_path, tile_id, patch_id, file_name)
print(s2)
os.path.exists(s2)
```
